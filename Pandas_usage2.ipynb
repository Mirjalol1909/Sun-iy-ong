```python
from numpy import polyadd
import pandas as pd

import requests

api_url = 'https://www.ncei.noaa.gov/cdo-web/api/v2/datacategories'
headers = {
    'token' : 'NcFrgbbBbhbTNLTFAHmmCcnqsmItgfOZ'
}
params = {
    'startdate': '1970-10-03',
    'enddate': '2012-09-10'
}
response = requests.get(api_url, headers=headers, params=params)
response.json()

```

```python
def make_request(endpoint, payload=None):
  return requests.get(
      f'https://www.ncei.noaa.gov/cdo-web/api/v2/{endpoint}',
      headers={
          'token' :'NcFrgbbBbhbTNLTFAHmmCcnqsmItgfOZ',
      },
      params=payload
  )
response2 = make_request
response.json()
```

```python
import datetime
from IPython import display

past = datetime.date(2026, 1, 1)
end = datetime.date(2026, 4, 1)

results = []

while past < end:
  display.clear_output(wait=True)
  display.display(f'Gathering data for {str(past)}')

  response2 = make_request(
      endpoint='data',
      payload={
          'datasetid': 'GHCND',
          'locationid' : 'CITY:US360019',
          'startdate': past,
          'enddate': past,
          'units': 'metric',
          'limit': 1000
      }
  )
  if response2.ok:
    results.extend(response2.json().get('results', []))
  past += datetime.timedelta(days=1)
```

```python
df = pd.DataFrame(results)
df.head(20)
```
