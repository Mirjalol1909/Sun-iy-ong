Min =**Maxscaler** [x' = x-min(x)/max(x)-min(x)]

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler



data = [[-1, 7], [-0.8, 2,], [-0.5, 6], [3, 9], [0.8, 18]]
min_max_scaler = MinMaxScaler().fit(data)
print(min_max_scaler.fit(data))
print()

z_scoring = StandardScaler()
print(z_scoring.fit(data))
print()

max_abs_scaler = MaxAbsScaler()
print(max_abs_scaler.fit(data))
print()

robust_scaler = RobustScaler().fit(data)
print(robust_scaler)
```

```python
print(min_max_scaler.data_max_)
print(min_max_scaler.data_min_)
print()

scaled = min_max_scaler.transform(data)
print(scaled)
print()

print(min_max_scaler.inverse_transform(scaled))
print()
```

Z-scoring [x' = (x-u)/y] normalizatsiya uchun

```python
print(z_scoring.mean_)
print()

print(z_scoring.transform(data))
print()
new_scaled = z_scoring.transform(data)
print(new_scaled)

print(z_scoring.inverse_transform(new_scaled))
```

MaxAbs Scoring [x' = x/Abs(max)]

```python
print(max_abs_scaler.transform(data))
print()

new_one = max_abs_scaler.transform(data)
print(max_abs_scaler.inverse_transform(new_one))
```

```python
print(robust_scaler.transform(data))
print()
new_robust = robust_scaler.transform(data)
print(robust_scaler.inverse_transform(robust_scaler.transform(data)))
```
